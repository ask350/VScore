## Two different functions
1. Calculates Hypotenuse for given sides of right angled triangle
2. Area of triangle for given base and height

```python
def hypotenuse(a,b):
    print('Hypotenuse function started')
    time.sleep(3)
    h = (a**2 + b**2)**(1/2)
    print(f'Hypotenuse for given sides {a} and {b} is {h:.4f}')
```

```python
def area(base, height):
    print('Area function started')
    time.sleep(3)
    a = (1/2)*base*height
    print(f"Area of given triangle is {a:.4f}")
```

```python
%%time
hypotenuse(3, 4)
```

```python
%%time
area(20, 10)
```

### Single Threaded Execution

```python
%%time
hypotenuse(12, 13) #3seconds
area(30, 14) #3seconds
```

### Multithreaded Execution

```python
from threading import Thread
```

```python
%%time
# Create the threads
th1 = Thread(target=hypotenuse, args=(13, 14))
th2 = Thread(target=area, args=(30, 20))

# Execute the threads
th1.start()
th2.start()

# wait for all threads to  finish
th1.join()
th2.join()
```

### Assign thread for one function but multiple values

```python
def square(n):
    time.sleep(2)
    s = n**2
    print(f"Square of given number {n} is {s}\n")
```

```python
%%time
square(100)
```

```python
%%time
square(16)
```

## Single threading

```python
%%time
a = [1, 2, 5, 11, 12]
for i in a:
    square(i)
```

### Multithreading assigning each thread a different number

```python
%%time
# Create thread
threads = []
for i in a:
    th = Thread(target = square, args=(i,))
    th.start()
    threads.append(th)


# Wait for all threads to complete
for th in threads:
    th.join()
```

```python
import numpy as np
nums = np.arange(1, 1000, 1)
nums
```

```python
%%time
# Create and start threads
threads = []
for i in nums:
    th = Thread(target=square, args=(i,))
    th.start()
    threads.append(th)

# Wait for all threads to finish
for th in threads:
    th.join()
```

```python
from urllib.request import urlretrieve  # function - downloads file
```

```python
def download_file(url):
    filename = url.split('/')[-1]
    print(f'Downloading : {filename} ....')
    urlretrieve(url, filename)
    print(f"{filename} Downloaded")
```

```python
download_file(urls[0])
```

```python
%%time
for i in urls:
    download_file(i)
```

# Multithreaded downloading

```python
%%time
# Create all the threads and start
threads = []
for i in urls:
    th = Thread(target=download_file, args=(i, ))
    th.start()
    threads.append(th)

# Wait for all threads to finish 
for th in threads:
    th.join()
```
