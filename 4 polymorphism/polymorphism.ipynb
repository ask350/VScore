# Triangle Assignment
Sum of any two sides should be greater than remaninig/3rd side
 


```python
class Triangle:

    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def isTriangle(self):
        a = self.a
        b = self.b
        c = self.c
        if (a+b>c) and (b+c>a) and (c+a>b):
            t = True
        else:
            t = False
        return t
    
    def perimeter(self):
        if self.isTriangle():
            p = self.a + self.b + self.c
        else:
            print(f'The given sides {self.a}, {self.b}, and {self.c} DO NOT FORM A Triangle')
            p = None
        return p
    
    def area(self):
        if self.isTriangle():
            s = (self.a + self.b + self.c)/2
            a = (s*(s-self.a)*(s-self.b)*(s-self.c))**(1/2)

        else:
            print(f'The given sides {self.a}, {self.b}, and {self.c} DO NOT FORM A TRIANGLE')
            a = None
        return a
```
