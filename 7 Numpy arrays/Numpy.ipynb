## Multiplying by 3 in a list vs array

```python
n = [3, 4, 5, 6, 7]
type(n)
```

```python
n*3     # here elements did not get mutiplyed but list was repeated thrice
```


### Always need to apply "for loop" for a list

```python
n2 =[]
for i in n:
    n2.append(i*3)
```

```python
n2
```

```python
m = np.array(n)
m
```

```python
m2 = 3*m           # called broadcasting in numpy array
m2
```

## Generating Normal Distribution

```python
np.random.normal(loc=5.5, scale=0.5/3)
```

```python
heights = np.random.normal(loc=5.5, scale=0.5/3, size=10000)
heights
```

```python
heights.mean()
```

```python
import seaborn as sns
sns.histplot(heights, kde=True)
```
